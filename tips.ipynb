#### 💡 LeetCode Tip: Regular Dict vs defaultdict

Use `defaultdict` to avoid key checking!

```python
from collections import defaultdict

# Regular dict - need to check before use
d = {}
d["a"] = 1
# d["b"] += 1  # Would raise KeyError!

# Need to check first
if "b" not in d:
    d["b"] = []
d["b"].append(1)
print(f"Regular dict: {d}")

# defaultdict - auto-creates with default value
d2 = defaultdict(int)
d2["a"] = 1
d2["b"] += 1  # Just works! Auto-creates with 0
print(f"\ndefaultdict(int): {dict(d2)}")

# defaultdict with list
d3 = defaultdict(list)
d3["a"].append(1)  # Auto-creates empty list
d3["a"].append(2)
print(f"defaultdict(list): {dict(d3)}")
```

### <a id='stack'></a>Stack

Queue-like structure, O(1) to all operations. Last-in-first-out; LIFO.

```python
from collections import deque

stk = []
print(stk)

#Append to top of stack - O(1)
stk.append(5)
print(stk)

stk.append(4)
stk.append(3)
print(stk)

# Pop from stack - O(1)
x = stk.pop()
print("x:",x)
print("stack after pop:", stk)

# Ask what's on the top - O(1)   bottom <-----stack-----> top 
print("On the top:",stk[-1])


# Ask if something is in the stack, wether it's empty - O(1)
if stk:
    print(True)

stk.pop() #removes the 4
stk.pop() #removes the 3
print("stack after popping all:",stk)

# If we try to check if something is in the stack now, it shall return an error
# because we removed all elements from it. Use a if stk check before using pop!!

if stk:
    print(True)
else:
    print(False)
```

### <a id='queues'></a>Queues (Deque)

Double-ended queue with O(1) operations at both ends. First-in-first-out; FIFO.

```python
from collections import deque

queue = deque()

# Enqueue - Add element to the right - O(1)
queue.append(1)
queue.append(2)
queue.append(3)
print(f"After appends: {queue}")

# Dequeue - Remove from the left - O(1)
queue.popleft()
print(f"After popleft(): {queue}")

# Add to front
queue.appendleft(0)
print(f"After appendleft(0): {queue}")

# Peek from left side - O(1)
print("First element:",queue[0])

#Peek from right side - O(1
print("Last element:", queue[-1])
```

### <a id='heaps'></a>Heaps (Priority Queue)

Min heap implementation. For max heap, negate values.

```python
import heapq

# Min Heap
minHeap = []
heapq.heappush(minHeap, 4)
heapq.heappush(minHeap, 3)
heapq.heappush(minHeap, 2)

print("Min Heap:")
print(f"Smallest element (root): {minHeap[0]}")
print(f"Elements in ascending order:")
while len(minHeap):
    print(f"  {heapq.heappop(minHeap)}")

print()

# Max Heap (negate values)
maxHeap = []
heapq.heappush(maxHeap, -4)
heapq.heappush(maxHeap, -3)
heapq.heappush(maxHeap, -2)

print("Max Heap (negated):")
print(f"Largest element (root): {-maxHeap[0]}")
print(f"Elements in descending order:")
while len(maxHeap):
    print(f"  {-heapq.heappop(maxHeap)}")

print()

# Building heap from array
arr = [5, 2, 9, 1, 5, 6]
heapq.heapify(arr)
print(f"Heapified array: {arr}")
print(f"Smallest element: {arr[0]}")
```

### Modifying Outer Variables with `nonlocal`

```python
def double(arr, val):
    def helper():
        # Can modify objects (lists) in place
        for i, n in enumerate(arr):
            arr[i] *= 2
        
        # Need nonlocal to reassign variables from outer scope
        nonlocal val
        val *= 2
    
    helper()
    print(f"Array after doubling: {arr}")
    print(f"Value after doubling: {val}")

nums = [1, 2]
val = 3
double(nums, val)
```

### <a id='two-pointers'></a>Two Pointers Pattern

**When to use:**
- Sorted arrays or can be sorted
- Finding pairs, triplets, or ranges
- Comparing elements from both ends

**Key concepts:**
- **1-indexed array**: Starts at index 1
- **Sorted in non-decreasing order**: Increasing but may have repeated values
- Move pointers based on comparisons to eliminate impossible solutions

**Common problems:**
- Two Sum (sorted array)
- Three Sum
- Container With Most Water
- Valid Palindrome

#### Example 1: Container With Most Water

**Problem:** Find two heights that form the largest container.

**Strategy:**
1. Start with widest possible container (left=0, right=end)
2. Calculate area = `min(height[left], height[right]) * (right - left)`
3. Move the pointer at the **shorter** height (greedy approach)
4. Update max area

**Why move the shorter pointer?**
- Width always decreases when we move inward
- Height is limited by the shorter side
- Moving the shorter pointer gives us a chance to find a taller height
- Moving the taller pointer can only decrease or maintain the area

**Time Complexity:** O(n)  
**Space Complexity:** O(1)

```python
def maxArea(height):
    """
    Find the container with the most water.
    
    Args:
        height: List of heights
    
    Returns:
        Maximum area
    """
    max_area = 0
    left = 0
    right = len(height) - 1
    
    while left < right:
        # Calculate current area
        width = right - left
        curr_area = min(height[left], height[right]) * width
        max_area = max(max_area, curr_area)
        
        # Move the pointer at the shorter height
        if height[left] < height[right]:
            left += 1
        else:
            right -= 1
    
    return max_area

# Test
print("Container With Most Water:")
test_cases = [
    [1, 8, 6, 2, 5, 4, 8, 3, 7],
    [8, 7, 2, 1],
    [1, 1]
]

for heights in test_cases:
    result = maxArea(heights)
    print(f"  heights={heights} → max_area={result}")
```

#### Example 2: Three Sum

**Problem:** Find all unique triplets that sum to zero.

**Strategy:**
1. **Sort** the array first
2. Fix one element with outer loop
3. Use **two pointers** on remaining elements
4. Skip duplicates to avoid duplicate triplets

**Key insights:**
- Sorting enables two-pointer approach and duplicate handling
- Skip duplicate values for the fixed element
- After finding a valid triplet, skip duplicates for both pointers
- Move both pointers after finding a match to continue searching

**Time Complexity:** O(n²) - outer loop O(n) × two-pointer O(n)  
**Space Complexity:** O(1) or O(log n) for sorting (excluding output)

```python
def threeSum(nums):
    """
    Find all unique triplets that sum to zero.
    
    Args:
        nums: List of integers
    
    Returns:
        List of triplets [a, b, c] where a + b + c = 0
    """
    nums.sort()  # Sort in place
    result = []
    
    for i in range(len(nums)):
        # Skip duplicates for the fixed element
        if i > 0 and nums[i] == nums[i - 1]:
            continue
        
        # Two pointers for the remaining array
        left = i + 1
        right = len(nums) - 1
        target = -nums[i]
        
        while left < right:
            current_sum = nums[left] + nums[right]
            
            if current_sum < target:
                left += 1
            elif current_sum > target:
                right -= 1
            else:
                # Found a valid triplet
                result.append([nums[i], nums[left], nums[right]])
                
                # Move both pointers and skip duplicates
                left += 1
                right -= 1
                
                while left < right and nums[left] == nums[left - 1]:
                    left += 1
                while left < right and nums[right] == nums[right + 1]:
                    right -= 1
    
    return result

# Test
print("Three Sum:")
test_cases = [
    [-1, 0, 1, 2, -1, -4],
    [0, 0, 0],
    [-2, 0, 1, 1, 2]
]

for nums in test_cases:
    result = threeSum(nums)
    print(f"  nums={nums}")
    print(f"  result={result}\n")
```

#### Two Pointers: Key Differences

| Problem | Needs Sorting? | Pointers | Why? |
|---------|---------------|----------|------|
| **3Sum** | ✅ Yes | 3 (1 fixed + 2 moving) | Need sorted values to compare sums |
| **Container With Most Water** | ❌ No | 2 (both moving) | Index positions matter for width |

**General pattern:**
- **n elements to compare** → Often needs **1 fixed + (n-1) pointers** or **n pointers**
- Two-pointer works best for **2-3 element problems** on sorted or position-based arrays
- Make binary decisions to eliminate impossibilities

### <a id='binary-search'></a>Binary Search

```python
# Define L, M, R indexes
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        #ologn calls for binary search
        l = 0
        r = len(nums) - 1
        m = (l + r) // 2

        while l <= r:
            if nums[m] == target:
                return m
            elif target > nums[m]:
                l = m + 1
                m = (l + r) // 2
            elif target < nums[m]:
                r = m - 1
                m = (l + r) // 2
        return -1

sol = Solution()
print(sol.search([-1,0,3,5,9,12],9))
print(sol.search([-1,0,3,5,9,12],0))
```

## Trees

Array representation: 

    i
2i     2i+1

DFS: Depth-first search -> Preorder, Inorder, Postorder
- can be implemented w/o recursion using stack

BFS: Breadth-first search -> Level order traversal
- can be implemented w/o recursion using queue/deque


```python
# Binary trees implementation

class TreeNode:
    def __init__(self, val, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

    def __str__(self): #redefining the print for this object
        return str(self.val) #returns the string of val
    
```

```python
#        1
#   2          3
# 4   5     10  

# above is the tree we'll be working on
```

```python
A = TreeNode(1)
B = TreeNode(2)
C = TreeNode(3)
D = TreeNode(4)
E = TreeNode(5)
F = TreeNode(10)

A.left = B
A.right = C
B.left = D
B.right = E
C.left = F

print(A)
print(C.left)
```

```python
# Recursive Pre-Order Traversal (DFS), Time: O(n), Space: O(n)

def pre_order(node):
    #preorder: node, left, right

    if not node:
        return

    print(node)
    pre_order(node.left)
    pre_order(node.right)

pre_order(A)
```

```python
# Recursive In-Order Traversal (DFS), Time: O(n), Space: O(n)

def in_order(node):
    #inorder: left, node, right

    if not node:
        return

    in_order(node.left)
    print(node)
    in_order(node.right)

in_order(A)
```

```python
# Recursive Post-Order Traversal (DFS), Time: O(n), Space: O(n)

def post_order(node):
    #postorder: left,right,node

    if not node:
        return

    post_order(node.left)
    post_order(node.right)
    print(node)

post_order(A)
```

```python
# Iterative Pre Order Traversal (DFS) Time: O(n), Space: O(n) - using stacks

def pre_order_iterative(node):
    stk = [node]

    while stk:
        node = stk.pop()
        print(node)
        if node.right: stk.append(node.right)
        if node.left: stk.append(node.left)

pre_order_iterative(A)
```

```python
# Level Order Traversal (BFS) Time: O(n), Space: O(n) - using stacks
from collections import deque

def level_order(node):
    q = deque()
    q.append(node)

    while q:
        node = q.popleft()
        print(node)
        if node.left: q.append(node.left)
        if node.right: q.append(node.right)

level_order(A)
```

```python
# Check if value exists (DFS) Time: O(n), Space: O(n)

def search(node, target):
    if not node:
        return False

    if node.val == target:
        return True

    return search(node.left, target) or search(node.right, target)

search(A,10)
```

```python
# Binary Search Trees (BSTs)

#         5
#    1         8
# -1   3     7  9

A2 = TreeNode(5)
B2 = TreeNode(1)
C2 = TreeNode(8)
D2 = TreeNode(-1)
E2 = TreeNode(3)
F2 = TreeNode(7)
G2 = TreeNode(9)

A2.left, A2.right = B2, C2
B2.left, B2.right = D2, E2
C2.left, C2.right = F2, G2

print(A2)
```

```python
in_order(A2)
```

```python
# Time: O(log n), Space: O(log n)
def search_bst(node, target):
    if not node:
        return False

    if node.val == target:
        return True

    if target < node.val: return search_bst(node.left, target)
    else: return search_bst(node.right, target)

search_bst(A2, 111)
```

- "Find minimum/maximum value that satisfies a condition"
- Search space has clear bounds (min to max)
- Monotonic property: if k works, all values > k (or < k) also work

**Key insight:** Binary search on the ANSWER, not on input data

**Common problems:**
- Koko Eating Bananas
- Capacity to Ship Packages
- Minimum Speed to Arrive on Time
- Split Array Largest Sum


```python
def binary_search_on_answer(data, target):
    # Define search space bounds
    left = minimum_possible_answer
    right = maximum_possible_answer
    
    while left <= right:
        mid = (left + right) // 2
        
        # Check if mid satisfies the condition
        if is_valid(mid):
            # mid works, try to optimize (find smaller/larger)
            right = mid - 1  # For minimum problems
        else:
            # mid doesn't work, adjust search
            left = mid + 1
    
    return left  # After loop, left points to first valid answer
```

```python
#### Example: Koko Eating Bananas
def minEatingSpeed(piles: List[int], h: int) -> int:
    left, right = 1, max(piles)
    
    while left <= right:
        k = (left + right) // 2
        
        # Calculate total time with speed k
        total_time = sum((pile + k - 1) // k for pile in piles)
        
        if total_time <= h:
            # Speed works, try slower (we want minimum)
            right = k - 1
        else:
            # Too slow, need faster
            left = k + 1
    
    return left
```
